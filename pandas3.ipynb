```python
import pandas as pd
```

```python
df1 = pd.read_csv("D:\IHME-GBD_2019_DATA-15798851-2.csv")
df1.head(10)
```

```python
df2 = pd.read_csv("D:\IHME-GBD_2019_DATA-ff08d9bc-1.csv")
df2.head(10)
```

```python
df1.rename(columns = {'val' : 'daly_val', 'upper' : 'daly_upper', 'lower' : 'daly_lower'}, inplace = True)
df2.rename(columns = {'val' : 'deaths_val', 'upper' : 'deaths_upper', 'lower' : 'deaths_lower'}, inplace = True)
```

```python
df_all = pd.merge(df1, df2, on = ['location', 'sex', 'age', 'cause', 'metric', 'year'], how = 'inner')
```

```python
df_all.shape
```

```python
df_all_1 = df_all.iloc[:50000,:]
df_all_2 = df_all.iloc[50000:,:]
```

```python
df_all_new = pd.concat([df_all_1, df_all_2], axis = 0)
df_all_new.shape
```

```python
df_all["deaths_val_round"] = df_all["deaths_val"].round(decimals=1)
```

```python
df_all
```

```python
df_all['deaths_total'] = df_all['deaths_val'] + df_all['deaths_upper'] + df_all['deaths_lower']
df_all
```

```python
df_all['total_loc'] = df_all.loc[:, 'deaths_val':'deaths_lower'].sum(axis=1)
df_all
```

```python
df_all['total_iloc'] = df_all.iloc[:, 11:13].sum(axis=1)
```

```python
df_all
```

```python
df_all['total_iloc'] = df_all.iloc(:, 11:14).sum(axis=1)
```

```python
df_all
```

```python
df_all.drop(['deaths_val_round', 'deaths_total', 'total_loc', 'total_iloc'], axis=1, inplace=True)
```

```python
CIS_2020 = ['Azerbaijan', 'Armenia', 'Belarus', 'Kazakhstan', 'Kyrgyzstan', 
            'Moldova', 'Russian Federation', 'Tajikistan', 'Uzbekistan']
```

```python
df_all['CIS_2020'] = df_all['location'].apply(lambda x: True if x in CIS_2020 else False)
df_all[df_all['CIS_2020'] == True]
```

```python
df_all['CIS_2020'] = df_all['location'].apply(lambda x: x in CIS_2020)
```

```python
def if_cis(x):
    if x in CIS_2020:
        return True
    else:
        False
```

```python
df_all['CIS_2020'] = df_all['location'].apply(if_cis)
```
